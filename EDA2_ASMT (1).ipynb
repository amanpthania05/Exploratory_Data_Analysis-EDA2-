```python
#Outlier Detection for all columns
Num_Columns= ['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']
for i in Num_Columns:
  column_name = i
  print("Outliers calculation for", i)
  print(adult.boxplot(column=[column_name],grid=True))
  print(adult[column_name].describe())
#Quartile 1
  print(adult[column_name].describe())
  Q1=adult[column_name].quantile(0.25)
  print('Quartile 1 : ',Q1)
#Quartile 3
  Q3=adult[column_name].quantile(0.75)
  print('Quartile 3 : ',Q3)
#Interquartile Range
  IQR= Q3 - Q1
  print('Inter-Quartile Range : ',IQR)
#Upper Limit
  UL = Q3 + (1.5 * IQR)
  print('Upper Limit : ',UL)
#Lower Limit
  LL = Q1 - (1.5 * IQR)
  print('Lower Limit : ', LL)
  Outliers=adult[(adult[column_name] > UL) | (adult[column_name] < LL)]
  print(Outliers)
  print('__________________________________________________________')
```

# **PPS SCORE**

```python
import ppscore as pps
```

```python
# ISOLATION FOREST
adultt_encoded=pd.get_dummies(adultt)
adultt_encoded
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(adultt_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(adultt_encoded)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

```python
adultt_encoded['Outliers']=y_pred_outliers
```

```python
adultt_encoded
```

```python
adultt_encoded[adultt_encoded.Outliers==-1]
```

```python
adultt_encoded.drop('Outliers', axis=1,inplace=True)   # DROPPING OUTLIERS
adultt_encoded
```

```python
pps.score(adultt_encoded, "age", "sex_ Male")
```
